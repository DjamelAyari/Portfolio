```python
import hashlib
```

```python
export_csv['Prénom/Nom'] = export_csv['Prénom/Nom'].apply(
    lambda x: 
        hashlib.sha256(x.encode()).hexdigest()
)
```

```python
export_csv['Telephone'] = export_csv['Telephone'].astype(str)
```

```python
export_csv['Telephone'] = export_csv['Telephone'].apply(
    lambda x: 
        hashlib.sha256(x.encode()).hexdigest()
)
```

```python
export_csv['Date_naissance'] = export_csv['Date_naissance'].astype(str)
```

```python
export_csv['Date_naissance'] = export_csv['Date_naissance'].apply(
    lambda x: 
        hashlib.sha256(x.encode()).hexdigest()
)
```

```python
export_csv['Enfants'] = export_csv['Enfants'].astype(str)
```

```python
export_csv['Enfants'] = export_csv['Enfants'].apply(
    lambda x: 
        hashlib.sha256(x.encode()).hexdigest()
)
```

```python
export_csv['id_salarié'] = export_csv['id_salarié'].apply(
    lambda x: 
        hashlib.sha256(x.encode()).hexdigest()
)
```
