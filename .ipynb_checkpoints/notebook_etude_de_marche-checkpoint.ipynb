### <font color = "red">Recherche de corrélations</font>

Toutes mes colonnes sont des float et aucune de mes colonnes n'est distribué normalement, je vais donc utiliser la Spearman correlation.

1 --> Complete positive correlation

0 --> No correlation

-1 --> Complete negative correlation

```python
coré = df.corr(method = "spearman")
#coré.to_csv('coré.csv')
coré.round(2)

def HIGHLIGHT_COLOR(x):
    def colour_switch(number):
        if number >= 0.50 and number <= 0.90:
            color = "green"
        elif number <= -0.50 and number >= -0.90:
            color = "blue"
        elif number >= 0.90:
            color = "red"
        elif number <= -0.90:
            color = "red"

        else:
            # default
            color = "white"
            
        return color

    return [f'background-color: {colour_switch(number)}' for number in x]

coré.style.apply(HIGHLIGHT_COLOR)

#There are NaN because data does not varies.
#Correlation formula --> cor(i,j) = cov(i,j)/[stdev(i)*stdev(j)]
```

## <center><font color = "yellow">PCA</font></center>

```python
from sklearn.decomposition import PCA

pcamodel = PCA(n_components = 11)
df_pca = pd.DataFrame(pcamodel.fit_transform(df_scaled),columns=df_scaled.columns,index = df_scaled.index)
print(df_pca.shape)
df_pca.head(5)
```

```python
pcamodel.explained_variance_ratio_
```

```python
pcamodel.components_
```

```python
explained_variance = pd.Series(dict(zip(df_pca.columns, 100.0*pcamodel.explained_variance_ratio_)))
print(explained_variance)
```

```python
explained_variance.plot(kind='bar', figsize=(15, 4), rot=90, ylabel='Explained variance')
```

```python
explained_variance['Disponibilité alimentaire (Kcal/personne/jour)'] + explained_variance['Disponibilité alimentaire en quantité (kg/personne/an)'] + explained_variance['Disponibilité de matière grasse en quantité (g/personne/jour)']
```

```python
explained_variance['Disponibilité alimentaire (Kcal/personne/jour)'] + explained_variance['Disponibilité de matière grasse en quantité (g/personne/jour)']
```
